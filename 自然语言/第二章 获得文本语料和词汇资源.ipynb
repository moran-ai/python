```python
# 比较不同文本中情态动词的用法
# 第一步，对特定文体进行计数,在之前输入nltk
from nltk.corpus import brown
news_text = brown.words(categories='news')
fdist = nltk.FreqDist([w.lower() for w in news_text])
modals = ['can','could','may','might','must','will']
for m in modals:
    print(m + ':',fdist[m],)
```

```python
# 统计每一个感兴趣的文体
cfd = nltk.ConditionalFreqDist(
           (genre, word)
           for genre in brown.categories()  # 分类
           for word in brown.words(categories='news'))  # 分类下单词出现的频率
```

```python
genres=['news','religion','hobbies','science_fiction','romance','humor']
modals=['can','could','may','might','must','will']
cfd.tabulate(conditions=genres, samples=modals) # 二维表格 tabulate, 分类conditions=genres, 数值samples=modals
```

```python
cfd.keys() # 打印所有的键 
```

```python
cfd['news']  # 新闻的键值
```

```python
cfd['news'].keys() # 出现了哪些单词
```

```python
cfd['news']['The'] # The单词出现的频率
```

```python
def my_tabulate(genesis,modals):
    # 打印若干空格，跟随modals列表
    genres=['news','religion','hobbies','science_fiction','romance','humor']
    modals=['can','could','may','might','must','will']
    for g in genesis:
        # 打印g
        for m in modals:
            print(cdf[g][m],end="")
            print('',end='')
```

```python
my_tabulate(['news','religion','hobbies'],['can','could','may'],cdf)
```
