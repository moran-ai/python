## 13.定义一个变量silly包含字符串：'newly formed bland ideas are inexpressible in an infuriating way'

### a分割silly为一个字符串链表，一个词一个链表，一个词一个字符串，使用python的split操作，并保存到bland的变量中

```python
silly='newly formed bland ideas are inexpressible in an infuriating way'
bland=silly.split()
bland
```

### b提取silly中每一个词的第二个字母，将他们连接成一个字符串，得到'eoldrnnnna'

```python
silly1=[]
for word in bland:
    silly1.append(word[1])
```

```python
silly1
```

```python
''.join(silly1)
```

### c 使用join()将bland中的词组合到一个单独的字符串，确保结果字符串中的词可以隔开

```python
newsilly=' '.join(bland)
newsilly
```

### d 按字母顺序输出silly中的词

```python
silly2 = sorted(set(silly1))
print(silly2)
```
