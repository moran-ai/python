# 正则表达式的有益应用

```python
import re
import nltk
word = 'supercalfragilisticexpiailidocious'
re.findall(r'[aeiou]',word)  # findall匹配所有指定的表达式
```

```python
len(re.findall(r'[aeiou]',word))
```

```python
# 找出文本中两个或两个以上的元音序列，并确定它们的相对频率
wsj = sorted(set(nltk.corpus.treebank.words()))
dfd = nltk.FreqDist(vs for word in wsj for vs in re.findall(r'[aeiou]{2,}',word))
dfd.items()
```

# 在字符块上做更多事情

```python
#使用 re.findall()提取所有匹配的词中的字符，然后使用''.join()将它们连接在一 起
import re
import nltk
regexp = r'^[AEIOUaeiou]+|[AEIOUaeiou]+$|[^AEIOUaeiou]'
def compress(word):
    pieces = re.findall(regexp, word)
    return "".join(pieces)

english_udhr = nltk.corpus.udhr.words("English-Latin1")
print(nltk.tokenwrap(compress(w) for w in english_udhr[:75]))
```

```python
rotokas_words = nltk.corpus.toolbox.words('rotokas.dic')
cvs = [cv for w in rotokas_words for cv in re.findall(r'[ptksvr][aeiou]',w)]
cfd = nltk.ConditionalFreqDist(cvs)
cfd.tabulate()
```

```python
cv_word_pairs = [(cv,w) for w in rotokas_words for cv in re.findall(r'[ptksvr][aeiou]',w)]
cv_index = nltk.Index(cv_word_pairs)
cv_index['su']
```

# 查找词干

```python
def stem(word):
    for suffix in ['ing','ly','ed','ious','ies','ive','es','s','ment']:
        if word.endswith(suffix):
                         return word[:-len(suffix)]
        return word
```

```python
re.findall(r'^.*(ing|ly|ed|ious|ies|ive|es|s|ment)$','processing')
```

```python
re.findall(r'^.*(?:ing|ly|ed|ious|ies|ive|es|s|ment)$','processing')
```

```python
re.findall(r'^(.*)(ing|ly|ed|ious|ies|ive|es|s|ment)$','processing')
```

```python
re.findall(r'^(.*)(ing|ly|ed|ious|ies|ive|es|s|ment)$','processes')
```

```python
re.findall(r'^(.*?)(ing|ly|ed|ious|ies|ive|es|s|ment)$','processes')
```

```python
re.findall(r'^(.*?)(ing|ly|ed|ious|ies|ive|es|s|ment)?$','language')
```

```python
def stem(word):
    regexp = r'^(.* ?)(ing|ly|ed|ious|ies|ive|es|s|ment)?$'
    stem,suffix = re.findall(regexp,word)[0]
    return stem
```
